```python
# Create a backup copy of the dataset to keep the original data safe
copy = df.copy()
copy.head()
```

```python
# Total Number of Guests
copy['total_guests'] = copy['no_of_adults'] + copy['no_of_children']   
```

```python
# Number of Nights
copy['total_nights'] = copy['no_of_weekend_nights'] + copy['no_of_week_nights']
```

```python
# Total booking cost = room price * total nights 
copy['total_booking_value'] = (copy['avg_price_per_room'] * copy['total_nights'])
```

```python
# Create function to make group for all months to 4 seasons
def season(month):

    if month in [12,1,2]:
        return "Winter"

    elif month in [3,4,5]:
        return "Spring"

    elif month in [6,7,8]:
        return "Summer"

    else:
        return "Autumn"

copy['season'] = copy['arrival_month'].apply(season)
```

```python
print(copy[['arrival_year','arrival_month','arrival_date']].head())
```

```python
copy["arrival_date_full"] = pd.to_datetime(
    {
        "year": copy["arrival_year"],
        "month": copy["arrival_month"],
        "day": copy["arrival_date"]
    },
    errors="coerce"
)
```

```python
# Classify bookings based on lead time: Greater than 60 days is early, otherwise late

copy['booking_type'] = np.where(copy['lead_time'] > 60,'Early','Late') 
```

```python
# --- Outlier Detection & Removal for Key Columns (IQR Method) ---

# Select only the 3 most important continuous numerical columns
important_num_cols = ['lead_time', 'avg_price_per_room', 'total_booking_value']

print("Rows BEFORE removing outliers:", copy.shape[0])

# Loop through each key column to filter out outliers
for col in important_num_cols:
    Q1 = copy[col].quantile(0.25)
    Q3 = copy[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define upper and lower bounds using the 1.5 * IQR rule
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Keep only the rows within the safe boundaries
    copy = copy[(copy[col] >= lower_bound) & (copy[col] <= upper_bound)]

print("Rows AFTER removing outliers:", copy.shape[0])
```

```python
# Encode Target Variable

copy['booking_status'] = copy['booking_status'].replace({'Canceled': 1, 'Not_Canceled': 0})

copy[['booking_status']].head(10)
```
